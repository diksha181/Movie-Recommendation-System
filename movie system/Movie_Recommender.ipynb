```python
#convert to list of strings for Genres and Keywords
import ast
def convert(obj):
  lst = []
  for i in ast.literal_eval(obj):                    ## (abstract syntax tree) converts the string to dict
    lst.append(i['name'])
  return lst

#convert to list of strings for cast
def convert2(obj):
  lst = []
  counter=0
  for i in ast.literal_eval(str(obj)):
    if counter!=3:
      lst.append(i['name'])
      counter+=1
    else:
      break
  return lst

#convert to list of strings for crew (director)
def convert3(obj):
  lst = []
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      lst.append(i['name'])
      break
  return lst
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split()) ##The split() method is used to split a string into a list of substrings based on a specified delimiter (by default, it splits on whitespace characters).
```

```python
movies.head()
```

```python
#removing spaces
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['overview'] = movies['overview'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#forming column by concatinating all these columns
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']
df = movies[['id','title','tags']]
df['tags']=df['tags'].apply(lambda x:" ".join(x))
```

```python
df['tags'][0]
```

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
df.head()
```

```python
#calculate similarity using text vectorization
#Technique used: Bag of Words - top 5000 words based on frequency
#used in the total text of tags column. A matrix of shape (5000,5000)
#is formed, with all the movies against the words, filling in the frequency
#of that particular word appearing in the movie tag. Vectorisation is performed
#on worsd other than stop words (is,the,a,are,from,etc.)
```

```python
#stemming used to get the base words from similar words like actor, actors.
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vectors = cv.fit_transform(df['tags']).toarray()
vectors
```

```python
cv.get_feature_names_out()
```

```python
#we calculate the cosine distance between two movies to recommend
#similar distance, euclidean distance(tip to tip of vector) is not a
#good measure for higher dimension. Cosine distance -> angle between vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)    ## returns a matrix with numbers between 0-1 indicating similarity,
```

```python
#enumerate the similarity matrix, considering movies form the second index i.e. 1, since first index belongs to the same movie
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index = df[df.title==movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(df.iloc[i[0]].title)
```

```python
recommend('Frozen')
```
